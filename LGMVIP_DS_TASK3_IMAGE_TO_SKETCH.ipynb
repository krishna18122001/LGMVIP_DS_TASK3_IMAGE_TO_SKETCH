```python
pic = cv2.imread("golden-retriever-dog.jpg")
```

```python
plt.imshow(pic[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
plt.imshow(pic)
plt.axis(False)
plt.show()
```

```python
RGB_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_pic)
plt.axis(False)
plt.show()
```

```python
grey_pic =cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
```

```python
invert_pic =cv2.bitwise_not(grey_pic)
```

```python
blur_pic =cv2.GaussianBlur(invert_pic, (111,111),0)
```

```python
invblur_pic =cv2.bitwise_not(blur_pic)
```

```python
sketch_pic = cv2.divide(grey_pic, invblur_pic, scale=256.0)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Orginal', size=18)
plt.imshow(RGB_pic)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Pencil Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_pic, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
